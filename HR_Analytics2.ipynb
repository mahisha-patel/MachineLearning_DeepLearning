```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train_data.csv')
```

```python
df.head()
```

```python
df.drop('employee_id', axis=1, inplace=True)
```

```python
df.info()
```

```python
df.education.fillna(0)
```

```python
df.drop('region', axis=1, inplace=True)
df.drop('gender', axis=1, inplace=True)
```

```python
df.drop('recruitment_channel', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['training_score'] = df['no_of_trainings']*df['avg_training_score']
df.drop('no_of_trainings', axis=1, inplace=True)
df.drop('avg_training_score', axis=1, inplace=True)
```

```python
df.head()
```

```python
# generate binary values using get_dummies
dum_df = pd.get_dummies(df)

# merge with main df bridge_df on key values
df = df.merge(dum_df)

df.drop('department', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.drop('education', axis=1, inplace=True)
```

```python
df = df.dropna()
```

```python
X = df.drop("is_promoted",axis = 1)
y = df["is_promoted"]
```

```python
X
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X, y)
y_pred = rf.predict(X)
f1_score(y, y_pred)
```
